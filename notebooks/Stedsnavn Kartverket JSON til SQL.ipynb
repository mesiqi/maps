```python
df = pd.DataFrame.from_dict(data['features'], orient='columns')
```

```python
df['navn'] = df['properties'].apply(lambda x: x['enh_snavn'])
df['navntype'] = df['properties'].apply(lambda x: x['enh_navntype'])
df['coords'] = df['geometry'].apply(lambda x: str(x['coordinates']).replace('[','').replace(']',''))
df['kommunenr'] = df['properties'].apply(lambda x: x['enh_komm'])
df['fylkenr'] = df['properties'].apply(lambda x: x['kom_fylkesnr'])
```

```python
df.drop(['geometry','properties'], axis=1, inplace=True)
```
